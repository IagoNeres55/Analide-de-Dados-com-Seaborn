```
import pandas as pd
import seaborn as sns
```

```
dataf = pd.read_csv("/content/Walmart.csv")
```

```
dataf.rename(columns={'Store':'Loja',
          'Date':'Data',
          'Weekly_Sales':'Vendas_Semanais','Holiday_Flag':'Feriados',
          'Temperature':'Temperatura','Fuel_Price':'Combustivel',
          'Unemployment':'Desemprego'}, inplace=True)
```

```
dataf.info()
```

```
dataf.describe()
```

**Contagem De Feriados e dê dias normais**

```
dataf['Feriados'].value_counts()
```

Quantidade de Vendas em Feriados

```
dataf[dataf['Feriados']==1]['Vendas_Semanais'].sum()
```

Quantidade de Vendas em dias Normais

```
dataf[dataf['Feriados']==0]['Vendas_Semanais'].sum()
```

Convertendo Data do nosso DataFrame

```
dataf['Mês'] = pd.to_datetime(dataf['Data'], format='%d-%m-%Y').dt.month
```

Plotando Gráfico de Barras, Qual mês ouve o maior número de vendas.

Novembro em feriados

Dezembro dias Normais

```
sns.barplot(data = dataf,
            x = 'Mês', 
            y = 'Vendas_Semanais',
            hue = 'Feriados',);
```

Verificar a alta do preço do combustível

```
 sns.lineplot(data=dataf,
              x='Mês',
              y='Combustivel',
              hue='Feriados');
```

Gráfico do tipo Scatter, Para visualizar o índice de desemprego comparado com o volume de vendas.

```
sns.scatterplot(data=dataf,
                x='Desemprego',
                y='Vendas_Semanais');
```
